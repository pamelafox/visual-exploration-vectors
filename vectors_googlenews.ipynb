```python
import gensim.downloader

google_news = gensim.downloader.load('word2vec-google-news-300')
# show the vector of the word 'computer'
print(google_news['computer'])
```

```python
# open most-common-nouns-english.csv and read the first column as words
import csv
import json

words = []
with open('most-common-nouns-english.csv') as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        words.append(row[0])

# for each word, get its vector
vectors = {}
for word in words:
    try:
        # turn np array into a python list
        vectors[word] = google_news[word].tolist()
    except KeyError:
        print(f'{word} not in the vocabulary')

# save the vectors to a file
with open('vectors_word2vec-google-news.json', 'w') as f:
    json.dump(vectors, f)
```
